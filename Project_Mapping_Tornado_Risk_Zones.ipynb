# Introduction

The zones identified as tornado risks face devastating natural storms that wreak violence at rates considered alarming by experts. According to III.ORG data 1,000 tornadoes are recorded yearly across the United States which cause more than 80 fatalities and 1,500 injuries annually (EN.WIKIPEDIA.ORG). Tornadoes produce winds exceeding 200 mph which effortlessly demolish regular houses and foundations extending to reinforced concrete structures (EN.WIKIPEDIA.ORG). Several part of the midwestern section known as Tornado Alley faces repeated tornado threats throughout each season which endangers numerous structures and their inhabitants (EN.WIKIPEDIA.ORG). Speedy tornado attacks together with catastrophic property destruction require communities to adopt effective safety measures in their high-risk areas to defend human lives and infrastructure. 

Strengthened vigilance exists as a fundamental necessity for communities dwelling in tornado-hit areas because they must reinforce the areas with high human concentrations. Tornado safety protocols should concentrate on strengthening protective elements for essential facilities which host large groups of people and critical infrastructure such as schools and hospitals as well as major transport routes. For example:

Schools: Children require fast shelter access against impending tornado danger at schools. The educational facilities that operate within tornado-prone regions run routine tornado emergency protocols to enable students along with personnel to access their protective areas immediately at any time (SPC.NOAA.GOV). Yearly repair of tornado safety drills enables students and staff to respond swiftly and calmly for shelter so children will stay safer in true tornado situations.

Hospitals: Medical facilities serve as crucial protective areas even during tornadoes since most patients need stable medical care due to their immobility and dependence on lifesaving equipment. Joplin Missouri witnessed an EF5 tornado strike the hospital and tear off roofing in 2011 which resulted in the loss of patient life support due to power failures (NASHVILLEMEDICALNEWS.COM). Tornado-prone areas recognize building strength and creating specific emergency shelters as essential hospital safety features. The hospital in Missouri created a steel reinforced shelter on its grounds that can safeguard 462 people including both staff and patients and community members through winds of up to 250 mph (TCMH.ORG). These fortified constructions protect every person at risk even when faced with violent storm conditions.

Highways: Drivers face emergency limitations when tornadoes hit during road travel since open highways provide minimal safety options. The wind speeds associated with tornados easily overpower vehicles as they provide no safety against winds (WEATHER.GOV notes that going under an overpass exposes people to fatal wind funnelling and flying debris). Major roads in areas prone to tornadoes now include traveler shelters as a measure to reduce the safety risks. Drivers who travel along the Kansas Turnpike can access metal storm shelters beneath toll facilities to seek safety during tornado alerts (CDLLIFE.COM). Getting into these roadside shelters becomes essential for safety since they offer protected areas during approaching tornadoes.


Populated Areas: Cities along with densely inhabited areas must deal with special difficulties when tornadoes strike their territories. When areas become densely populated this intensifies the danger of losing both many lives and extensive property damage. The protection of residents in tornado-prone urban settings features reinforced strong buildings and beneath-ground protection facilities together with emergency warning systems during severe weather events. The Oklahoma City administration has positioned public storm shelters across the city so residents missing basements can find protection in tornado alerts. Local authorities operate emergency broadcasting systems which combined with siren alerts give residents vital warning time that enables them to safely move to safety. The allocation of funding for these preparedness measures makes large urban residents safer against tornado threats.

The measures to recognize tornado risk zones together with implementing safety protocols in these areas serve a substantial necessity. Proactive preparations that include solid hospital designs and school drills and highway shelters allow communities to lower the threat posed by tornadoes to people and vital infrastructure. Our capability to survive harsh weather conditions remains stronger when we make our most exposed areas resistant and develop emergency contingency plans.

```python
# Load the shapefile
gdf = tornado_gdf_main

# Use 'yr' for the year and 'st' for state.
max_year = gdf['yr'].max()  # Should be 2023 given the dataset name

# most recent 1 year (only events in the most recent year)
past_1_year = gdf[gdf['yr'] == max_year].groupby('st').size().reset_index(name='Count')

# Past 5 years (from max_year-4 up to max_year)
past_5_years = gdf[gdf['yr'] >= (max_year - 4)].groupby('st').size().reset_index(name='Count')

# Past 10 years (from max_year-9 up to max_year)
past_10_years = gdf[gdf['yr'] >= (max_year - 9)].groupby('st').size().reset_index(name='Count')

# Table for 2010-2020 (inclusive)
table_2010_2020 = gdf[(gdf['yr'] >= 2010) & (gdf['yr'] <= 2020)].groupby('st').size().reset_index(name='Count')

# Table for 2000-2009 (using 2000 to 2009 so as not to overlap with the 2010-2020 period)
table_2000_2009 = gdf[(gdf['yr'] >= 2000) & (gdf['yr'] < 2010)].groupby('st').size().reset_index(name='Count')

# Table for 1990-1999 (using 1990 to 1999)
table_1990_1999 = gdf[(gdf['yr'] >= 1990) & (gdf['yr'] < 2000)].groupby('st').size().reset_index(name='Count')

# Print the tables
print("Tornado Counts by State - Past 1 Year (", max_year, "):")
print(past_1_year.to_string(index=False))
print("\nTornado Counts by State - Past 5 Years (", max_year-4, "to", max_year, "):")
print(past_5_years.to_string(index=False))
print("\nTornado Counts by State - Past 10 Years (", max_year-9, "to", max_year, "):")
print(past_10_years.to_string(index=False))
print("\nTornado Counts by State - 2010 to 2020:")
print(table_2010_2020.to_string(index=False))
print("\nTornado Counts by State - 2000 to 2009:")
print(table_2000_2009.to_string(index=False))
print("\nTornado Counts by State - 1990 to 1999:")
print(table_1990_1999.to_string(index=False))
```

Seasonal Tornado Analysis

```python
# Load the shapefile
gdf = tornado_gdf_main

# Create a new column 'season' based on the month (from 'mo')
def assign_season(month):
    if month in [1, 2, 3]:
        return 'Jan-Mar'
    elif month in [4, 5, 6]:
        return 'Apr-Jun'
    elif month in [7, 8, 9]:
        return 'Jul-Sep'
    elif month in [10, 11, 12]:
        return 'Oct-Dec'
    else:
        return 'Unknown'

gdf['season'] = gdf['mo'].apply(assign_season)

# Get the maximum year in the dataset (expected to be 2023)
max_year = gdf['yr'].max()

# Define a function to create and print a seasonality pivot table
def seasonality_table(df, title):
    # Group by state and season, then count events
    grouped = df.groupby(['st', 'season']).size().reset_index(name='Count')
    # Pivot so that states are rows and seasons are columns. Missing values become 0.
    pivot = grouped.pivot(index='st', columns='season', values='Count').fillna(0).astype(int)
    print(f"\nSeasonality Table for {title}:")
    print(pivot)
    return pivot


# Most Recent Year
df_past1 = gdf[gdf['yr'] == max_year]
table_past1 = seasonality_table(df_past1, f"Past 1 Year ({max_year})")

# Past 5 Years: events from (max_year-4) through max_year.
df_past5 = gdf[gdf['yr'] >= (max_year - 4)]
table_past5 = seasonality_table(df_past5, f"Past 5 Years ({max_year-4} to {max_year})")

# Past 10 Years: events from (max_year-9) through max_year.
df_past10 = gdf[gdf['yr'] >= (max_year - 9)]
table_past10 = seasonality_table(df_past10, f"Past 10 Years ({max_year-9} to {max_year})")

# --- Specific Time Periods ---

# 2010-2020 (inclusive)
df_2010_2020 = gdf[(gdf['yr'] >= 2010) & (gdf['yr'] <= 2020)]
table_2010_2020 = seasonality_table(df_2010_2020, "2010-2020")

# 2000-2009 (ensuring no overlap with 2010-2020)
df_2000_2009 = gdf[(gdf['yr'] >= 2000) & (gdf['yr'] < 2010)]
table_2000_2009 = seasonality_table(df_2000_2009, "2000-2009")

# 1990-1999
df_1990_1999 = gdf[(gdf['yr'] >= 1990) & (gdf['yr'] < 2000)]
table_1990_1999 = seasonality_table(df_1990_1999, "1990-1999")
```

```python
# Load the shapefile
gdf = tornado_gdf_main

# Define a function to assign a season based on the month (from the 'mo' column)
def assign_season(month):
    if month in [1, 2, 3]:
        return 'Jan-Mar'
    elif month in [4, 5, 6]:
        return 'Apr-Jun'
    elif month in [7, 8, 9]:
        return 'Jul-Sep'
    elif month in [10, 11, 12]:
        return 'Oct-Dec'
    else:
        return 'Unknown'

# Create a new column 'season' in the GeoDataFrame
gdf['season'] = gdf['mo'].apply(assign_season)

# Determine the maximum year (expected to be 2023 based on the dataset name)
max_year = gdf['yr'].max()

# Function to aggregate season counts and print a summary table
def aggregate_seasons(df, title):
    # Group by season and count the number of tornado events
    season_counts = df.groupby('season').size().reset_index(name='Count')
    # Order the seasons in a logical order
    season_order = ['Jan-Mar', 'Apr-Jun', 'Jul-Sep', 'Oct-Dec']
    season_counts['season'] = pd.Categorical(season_counts['season'], categories=season_order, ordered=True)
    season_counts = season_counts.sort_values('season')
    
    print(f"\nSeason Totals for {title}:")
    print(season_counts.to_string(index=False))
    return season_counts


# Past 1 Year: events in the most recent year
df_past1 = gdf[gdf['yr'] == max_year]

# Past 5 Years: events from (max_year - 4) through max_year
df_past5 = gdf[gdf['yr'] >= (max_year - 4)]

# Past 10 Years: events from (max_year - 9) through max_year
df_past10 = gdf[gdf['yr'] >= (max_year - 9)]

# Specific Time Periods:
df_2010_2020 = gdf[(gdf['yr'] >= 2010) & (gdf['yr'] <= 2020)]
df_2000_2009 = gdf[(gdf['yr'] >= 2000) & (gdf['yr'] < 2010)]
df_1990_1999 = gdf[(gdf['yr'] >= 1990) & (gdf['yr'] < 2000)]

# --- Aggregate and display season totals for each period ---

agg_past1    = aggregate_seasons(df_past1, f"Past 1 Year ({max_year})")
agg_past5    = aggregate_seasons(df_past5, f"Past 5 Years ({max_year-4} to {max_year})")
agg_past10   = aggregate_seasons(df_past10, f"Past 10 Years ({max_year-9} to {max_year})")
agg_2010_2020= aggregate_seasons(df_2010_2020, "2010-2020")
agg_2000_2009= aggregate_seasons(df_2000_2009, "2000-2009")
agg_1990_1999= aggregate_seasons(df_1990_1999, "1990-1999")
```
